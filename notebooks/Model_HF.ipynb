```python
from huggingface_hub import create_repo

repo_id = "NithinKeshav/fine-tuned-text2sql-mistral-7b"
try:
    create_repo(repo_id, repo_type="model", private=False)  # Set private=True if you want a private repo
    print(f"Repository {repo_id} created successfully!")
except Exception as e:
    print("Error creating repository:", e)
```

```python
from huggingface_hub import upload_folder

# Path to your model directory
model_path = "/content/drive/Shareddrives/DATA266_Project/Checkpoints/mistral_merged_full_4090"
repo_id = "NithinKeshav/fine-tuned-text2sql-mistral-7b"

# Upload the entire folder
try:
    upload_folder(
        repo_id=repo_id,
        folder_path=model_path,
        repo_type="model",
        commit_message="Upload fine-tuned Mistral 7B model for SQL generation"
    )
    print(f"Model successfully uploaded to {repo_id}!")
except Exception as e:
    print("Error uploading model:", e)
```
